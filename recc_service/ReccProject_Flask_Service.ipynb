```python
import ast
```

```python
# obj is basically: List of string objects:[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# ast.literal converts this string of objects into numeric indices.
# Then what we are doing is for each list of obj we are appending the name of tto the list. And the final list L will give us the genre of a particular movie.
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def convertCast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
def recommend_movies(parameter, values, top_n=5):
    """
    Recommend top N movies based on the selected parameter and values.

    :param parameter: The parameter to filter by (e.g., "Genre").
    :param values: A single value or list of values to filter by (e.g., "Action" or ["Action", "Drama"]).
    :param top_n: Number of top recommendations to return (default is 5).
    :return: DataFrame with top N movie recommendations.
    """
    if parameter == "Genre":
        # Handle multiple genres
        if isinstance(values, str):
            values = [values]

        # Transform selected genres into a feature vector that matches the model's expectations
        transformed_genres = mlb_genre.transform([values])  # Transform genres using MultiLabelBinarizer

        # Predict the cluster for the transformed genre feature vector
        cluster_ids = genre_birch.predict(transformed_genres)
        recommendations = movies_new[movies_new["cluster_genre"].isin(cluster_ids)]

    elif parameter == "Actor":
        # Handle multiple actors (if required)
        if isinstance(values, str):
            values = [values]
         # Transform selected actors into a feature vector
        transformed_actors = mlb_actor.transform([values])  # Assume mlb_actor is the MultiLabelBinarizer for actors

        # Predict the cluster for the transformed actor feature vector
        cluster_ids = actor_birch.predict(transformed_actors)
        recommendations = movies_new[movies_new["cluster_actor"].isin(cluster_ids)]

    elif parameter == "Director":
       if isinstance(values, str):
            values = [values]
        # Transform selected actors into a feature vector
       transformed_directors = mlb_director.transform([values])  # Assume mlb_actor is the MultiLabelBinarizer for actors

        # Predict the cluster for the transformed actor feature vector
       cluster_ids = director_birch.predict(transformed_directors)
       recommendations = movies_new[movies_new["cluster_director"].isin(cluster_ids)]

    elif parameter == "Release Year":

        # cluster_id = assign_cluster(values)
        # recommendations = movies_new[movies_new["cluster_year"]==cluster_id]
        if isinstance(values, int):  # If it's a single year
            cluster_id = assign_cluster(values)  # Single year input to assign_cluster
            recommendations = movies_new[movies_new["cluster_year"] == cluster_id]
        else:
            return "Invalid input for Release Year"

    else:
        return "Invalid parameter"

    # Select the top N movies
    return recommendations.head(top_n)
```

```python
import pandas as pd

def convert_to_frontend_format(dataframes_list):
    if not isinstance(dataframes_list, list):
        print("convert_to_frontend_format received non-list data.")
        return []

    # print(dataframes_list)
    formatted_movies = []

    for df in dataframes_list:
        if isinstance(df, pd.DataFrame):  # Process only valid DataFrames
            if not df.empty:
                formatted_movies.extend([
                    {
                        "title": row.get("title", "N/A"),
                        "year": row.get("release_year", "N/A"),
                        "rating": f"{row.get('rating', 0):.1f}" if "rating" in row else "N/A",
                        "image": row["image"] if pd.notna(row.get("image")) else "default.jpg",
                        "cast": row.get("cast", []),  # Actor list
                        "crew": row.get("crew", []),  # Director list
                        "description": " ".join(row.get("summary", [])) if isinstance(row.get("summary", ""), list) else row.get("summary", "N/A")
                    }
                    for _, row in df.iterrows()
                ])
        else:
            print(f"Skipping non-DataFrame object in dataframes_list: {type(df)}")

    return formatted_movies
```

```python
def store_reccomendations(user_preferences):
    reccomendations = {}
    for key, value in user_preferences.items():
        reccomendations[key] = []
        for i in value:
            # Capture the output of recommend_movies
            if key=="Year":
              key2="Release Year"
            else:
              key2=key
            print(key2)
            print(i)
            result = recommend_movies(key2,i, top_n=5)
            if isinstance(result, pd.DataFrame):  # Ensure only DataFrames are appended
                reccomendations[key].append(result)
            else:
                print(f"Unexpected result type from recommend_movies for {key}: {type(result)}")

    movies = {
        "genre": convert_to_frontend_format(reccomendations.get('Genre', [])),
        "actor": convert_to_frontend_format(reccomendations.get('Actor', [])),
        "director": convert_to_frontend_format(reccomendations.get('Director', [])),
        "year": convert_to_frontend_format(reccomendations.get('Year', []))
    }
    return movies

```

```python
import pandas as pd

def convert_series_to_frontend_format(series_list):
    """
    Converts a list of pandas Series objects into the desired frontend format.

    Parameters:
        series_list (list): A list of pandas Series objects containing movie data.

    Returns:
        list: A list of dictionaries formatted for the frontend.
    """
    if not isinstance(series_list, list):
        print("Input is not a list.")
        return []

    formatted_movies = []
    for series in series_list:
        if isinstance(series, pd.Series):
            formatted_movies.append({
                "title": series.get("title", "N/A"),
                "year": str(series.get("release_year", "N/A")),
                "rating": "N/A",  # Default as no 'rating' key is provided in the series.
                "image": series["image"] if pd.notna(series.get("image")) else "default.jpg",
                "cast": ", ".join(series.get("cast", [])),  # Convert cast list to comma-separated string.
                "crew": ", ".join(series.get("crew", [])),  # Convert crew list to comma-separated string.
                "description": series.get("summary", "N/A")  # Use summary as description.
            })
        else:
            print(f"Skipping non-Series object in series_list: {type(series)}")

    return formatted_movies
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

#Encode genres using MultiLabelBinarizer
mlb = MultiLabelBinarizer()
genre_vectors = mlb.fit_transform(movies_df['genres'])
genre_feature_names = mlb.classes_  # ['action', 'adventure', 'animation', ...]

# Vectorize summaries using TF-IDF
vectorizer = TfidfVectorizer()
summary_vectors = vectorizer.fit_transform(movies_df['summary']).toarray()

# Combine genres and summary features
movies_df['features'] = [list(genre_vec) + list(summary_vec)
                         for genre_vec, summary_vec in zip(genre_vectors, summary_vectors)]

# Create a dictionary to map movie titles to their features
movie_features = dict(zip(movies_df['title'], movies_df['features']))
```

```python
# Calculate similarity between user movies
def calculate_similarity(user1_movies, user2_movies):
    matched_count = 0
    total_count = len(user1_movies) + len(user2_movies)

    for movie1 in user1_movies:
        for movie2 in user2_movies:
            if movie1 in movie_features and movie2 in movie_features:
                sim_score = cosine_similarity(
                    movie_features[movie1],
                    movie_features[movie2]
                )[0, 0]
                if sim_score > 0.7:  # Consider a match if similarity > 0.7
                    matched_count += 1
    return (matched_count / total_count) * 100
```

```python
user1 = ["Toy Story", "Pinnochio"]
user2 = ["Toy Story", "Jumanji"]

similarity_percentage = calculate_similarity(user1, user2)
print(f"Similarity between User 1 and User 2: {similarity_percentage:.2f}%")
```
